```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style des graphiques
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# 1. CHARGEMENT DES DONNÉES
df = pd.read_csv('../data/data.csv', encoding='ISO-8859-1')
```

```python
print(f" AVANT nettoyage : {df.shape[0]} lignes")
```

```python
df.drop_duplicates(inplace=True)
print(f" Doublons supprimés: ")
```

```python
# 2. SUPPRIMER LES CUSTOMERID MANQUANTS
# On en a besoin pour les KPIs clients
df.dropna(subset=['CustomerID'], inplace=True)
print(f" CustomerID manquants supprimés :")
```

```python
# 3. VÉRIFIER LES VALEURS NÉGATIVES
print(f"\n Quantity négatives : {(df['Quantity'] < 0).sum()}")
print(f" UnitPrice négatives : {(df['UnitPrice'] <= 0).sum()}")
```

```python
# 4. GARDER UNIQUEMENT LES VENTES VALIDES
#  Quantity < 0 = retours / annulations
#  UnitPrice <= 0 = erreurs
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]

print(f"\n APRÈS nettoyage : {df.shape[0]} lignes")
```

```python
# 5. CRÉATION DE LA COLONNE CHIFFRE D'AFFAIRES
df['CA'] = df['Quantity'] * df['UnitPrice']

print(f"\n Colonne CA créée !")
print(df[['Quantity', 'UnitPrice', 'CA']].head(10))
```

```python
# 6. CONVERTIR LA DATE
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Mois'] = df['InvoiceDate'].dt.to_period('M')
df['Jour'] = df['InvoiceDate'].dt.date
df['Heure'] = df['InvoiceDate'].dt.hour

print(f" Période : {df['InvoiceDate'].min()} → {df['InvoiceDate'].max()}")
```

```python
# 7. RÉSUMÉ FINAL
print(" RÉSUMÉ APRÈS NETTOYAGE")
print(f"Lignes          : {df.shape[0]}")
print(f"Colonnes        : {df.shape[1]}")
print(f"Clients uniques : {df['CustomerID'].nunique()}")
print(f"Pays            : {df['Country'].nunique()}")
print(f"Produits        : {df['StockCode'].nunique()}")
print(f"Commandes       : {df['InvoiceNo'].nunique()}")
print(f"CA Total        : £{df['CA'].sum():,.2f}")
print(f"Valeurs manq.   : {df.isnull().sum().sum()}")
```

```python
df.to_csv('../data/clean_data.csv', index=False)
```
